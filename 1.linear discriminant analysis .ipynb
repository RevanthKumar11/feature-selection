```python
import pandas as pd
```

```python
dataSet = pd.read_csv('CKD.csv')
```

```python
dataSet.head()
```

```python
dataSet['classification'].value_counts()
```

```python
dataSet.loc[dataSet['classification'] == 'ckd', 'classification'] = 'yes' 
dataSet.loc[dataSet['classification'] == 'ckd\t', 'classification'] = 'yes' 
dataSet.loc[dataSet['classification'] == 'notckd', 'classification'] = 'no' 
```

```python
dataSet['classification'].value_counts()
```

```python
dataSet = pd.get_dummies(dataSet, drop_first = True)
```

```python
dataSet.head()
```

```python
independent = dataSet.drop('classification_yes', 1).values
dependent = dataSet['classification_yes'].values
```

```python
independent
```

```python
dependent
```

# split the train and test

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(independent, dependent, test_size = 0.25)
```

# standard scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xTrain = sc.fit_transform(xTrain)
xTest = sc.transform(xTest)
```

# linear discriminant analysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis(n_components = 3)
xTrain = lda.fit_transform(xTrain, yTrain)
xTest = lda.transform(xTest)
```
